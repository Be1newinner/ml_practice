```python
import pandas as pd
import numpy as np
```

```python
car_sales_csv = pd.read_csv("./data/car-sales-extended.csv")
car_sales_csv.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error 
```

```python
categorical_feature = ["Make", "Colour", "Doors"]
one_hot = OneHotEncoder(sparse_output=False)
transformer = ColumnTransformer([("one_hot", one_hot, categorical_feature)], remainder="passthrough") #we can use passthrough or drop for rest of the columns not in categorical feature
```

```python
x_data = car_sales_csv.drop("Price", axis=1)
y_train = car_sales_csv["Price"]
```

```python
x_data.head()
```

```python
y_train
```

```python
x_train = transformer.fit_transform(x_data)
```

```python
x_train
```

```python
dummies = pd.get_dummies(car_sales_csv[["Make", "Colour", "Doors"]][0:5])
dummies
```

```python
np.random.seed(42)
X_Train, X_Test, Y_Train, Y_Test = train_test_split(x_train, y_train, test_size=0.2)
model = RandomForestRegressor()
model.fit(X_Train, Y_Train)
```

```python
model.score(X_Test, Y_Test)
```

```python
pr_x_test = model.predict_prob(X_Test)
# pr_x_test.mean()
```

```python
self_x_test = np.array(Y_Test)
self_x_test.mean()
```

```python
mean_absolute_error(self_x_test, pr_x_test)
```
